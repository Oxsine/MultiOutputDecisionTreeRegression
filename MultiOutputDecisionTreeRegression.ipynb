```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

```python
# random seed
rng = np.random.RandomState(0)
#  sorted 2-dimensionl array from -100 to 100
X = np.sort(200 * rng.rand(100, 1) - 100, axis=0)
# array 100 * 2
# y[0] = cosX * pi
# y[1] =sinX * pi 
y = np.array([np.pi * np.cos(X).ravel(), np.pi * np.sin(X).ravel()]).T
# noise every 5th element
y[::5, :] += 0.5 - rng.rand(20, 2)
```

```python
# build the models with dif. max depth
regr_1 = DecisionTreeRegressor(max_depth = 2)
regr_2 = DecisionTreeRegressor(max_depth = 5)
regr_3 = DecisionTreeRegressor(max_depth = 8)
# filling the models
regr_1.fit(X, y)
regr_2.fit(X, y)
regr_3.fit(X, y)
```

```python
# Validational X
X_test = np.arange(-100, 100, 0.01)[:, np.newaxis]
# predicts
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
y_3 = regr_3.predict(X_test)
```

```python
plt.figure()
# the graph turns out to be a circle since 
# the cosine and sin of the angle are x and y points on the unit circle
plt.scatter(y[:, 0], y[:, 1], color = 'navy', s = 25, edgecolor = 'black', label  ='data')
plt.scatter(y_1[:, 0], y_1[:, 1], c = 'blue', s = 25, edgecolor = 'black', label = 'max_depth = 2')
plt.scatter(y_2[:, 0], y_2[:, 1], c = 'red', s = 25, edgecolor = 'black', label = 'max_depth = 5')
plt.scatter(y_3[:, 0], y_3[:, 1], c = 'orange', s = 25, edgecolor = 'black', label = 'max_depth = 8')
plt.legend(loc = 'best')
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.xlabel('target 1')
plt.xlabel('target 2')
plt.title("Multi-output Decision Tree Regression")
```
